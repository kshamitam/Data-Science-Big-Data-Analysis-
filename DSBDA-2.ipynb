```python
def fun1(value):
    if (value == "typical"):
        return 1
    elif (value == "asymptomatic"):
        return 2
    elif (value == "nonanginal"):
        return 3
    elif (value == "nontypical"):
        return 4
    else:
        pass
```

```python
heart["ChestPain"].value_counts()
```

```python
heart["ChestPain"] = heart["ChestPain"].apply(fun1)
```

```python
heart.head(10)
```

```python
def fun2(value):
    if (value == "fixed"):
        return 1
    elif (value == "normal"):
        return 2
    elif (value == "reversable"):
        return 3
    else:
        pass
```

```python
heart["Thal"].value_counts()
```

```python
heart["Thal"] = heart["Thal"].apply(fun2)
```

```python
heart.head(10)
```

```python
def fun3(value):
    if (value == "Yes"):
        return 1
    elif (value == "No"):
        return 0
    else:
        pass
```

```python
heart['AHD'] = heart["AHD"].apply(fun3)
```

```python
for x in ['RestBP']:
    q75,q25 = np.percentile(heart.loc[:,x],[75,25])
    intr_qr = q75-q25
 
    max = q75+(1.5*intr_qr)
    min = q25-(1.5*intr_qr)
 
    heart.loc[heart[x] < min,x] = np.nan
    heart.loc[heart[x] > max,x] = np.nan
```
